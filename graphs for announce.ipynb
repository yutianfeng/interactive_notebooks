```python
class PDF(object):
  def __init__(self, pdf, size=(200,200)):
    self.pdf = pdf
    self.size = size

  def _repr_html_(self):
    return '<iframe src={0} width={1[0]} height={1[1]}></iframe>'.format(self.pdf, self.size)

  def _repr_latex_(self):
    return r'\includegraphics[width=1.0\textwidth]{{{0}}}'.format(self.pdf)
```

```python
fig, ax1 = plt.subplots()

color = 'tab:red'
ax1.set_xlabel('Genome Position (bp)')
ax1.set_ylabel('Cumulative C-G Difference', color=color)
ax1.plot(readin[0], readin[1], color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:blue'
ax2.set_ylabel('C-G Skew in a 500Kbp Rolling Window', color=color)  # we already handled the x-label with ax1
ax2.plot(readin2[0], readin2[1], color=color)
ax2.tick_params(axis='y', labelcolor=color)

plt.show()
fig.savefig("strandbias.pdf", bbox_inches="tight", dpi=270)
```
